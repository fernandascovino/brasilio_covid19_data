## Dados do painel IndicaSUS - coleta em 13/out

```python
df = pd.read_excel("TabelaDinamica-13_10_2020_08_27_11.xlsx", header=2)
df.info()
```

### Filtra datas anteriores a março

```python
meses_validos = [f"0{i}/2020" if i < 10 else f"{i}/2020" for i in range(3, 11)]
meses_validos
```

```python
drop = df[(df["DataNotificacao"].str.contains('|'.join(meses_validos)) == False)
         | (df["DataModificacaoObito"].str.contains('|'.join(meses_validos)) == False)]
drop
```

```python
df_clean = df.drop(drop.index)
# drop.to_csv("br_mt_casos_removidos.csv")
```

```python
for col in ["DataNotificacao", "DataModificacaoObito", "DataEncerramentoCaso"]:
    df_clean[col] = pd.to_datetime(df[col], format="%d/%m/%Y %H:%M:%S")
```

```python
df_clean["DataNotificacao"].dt.month.value_counts().sort_index()
```

## Comparacao dos dados manuais x IndicaSUS

```python
dates = ['2020-09-17', '2020-09-02']

microdados = {date: pd.DataFrame() for date in dates}

# pega confirmados e óbitos para cada data
for date in dates:
    
    # Óbitos
    microdados[date]["Óbitos"] = (df_clean.query(f"DataModificacaoObito <= '{date}'")
#                                        .query("Situacao == 'Óbito'")
                                       .groupby("Municipio")["Seq_Row_Number"].count())

    # Casos
    microdados[date]["Confirmados"] = (df_clean.query(f"DataNotificacao <= '{date}'")
                                          .groupby("Municipio")["Seq_Row_Number"]
                                          .count())
    
    # Completa dias sem casos/obitos
    microdados[date] = microdados[date].fillna(0)


pd.options.display.max_rows=150
microdados['2020-09-17'] = microdados['2020-09-17'].join(
    compare_20200917.set_index("Municípios")[["Óbitos", "Confirmados"]], 
    lsuffix="_microdados",rsuffix="_manual")

microdados['2020-09-02'] = microdados['2020-09-02'].join(
    compare_20200902.set_index("Municípios")[["Óbitos", "Confirmados"]], 
    lsuffix="_microdados",rsuffix="_manual")
```

```python
# microdados['2020-09-17'].to_csv("br_mt_microdados_comparacao_20200917.csv")
# microdados['2020-09-02'].to_csv("br_mt_microdados_comparacao_20200902.csv")
```

## Serie historica acumulada - IndicaSUS

```python
confirmados = (df_clean
 .groupby(["Municipio", "DataNotificacao"])["Seq_Row_Number"]
 .count()
 .reset_index()
 .assign(confirmados= lambda df: df.groupby(["Municipio"])["Seq_Row_Number"].cumsum())
 .sort_values(["DataNotificacao", "Municipio"])
 .rename(columns={"DataNotificacao": "data", 
                  "Municipio": "municipio"})
 .drop(columns="Seq_Row_Number"))

confirmados
```

```python
mortes = (df_clean
 .groupby(["Municipio", "DataModificacaoObito"])["Seq_Row_Number"]
 .count()
 .reset_index()
 .assign(mortes= lambda df: df.groupby(["Municipio"])["Seq_Row_Number"].cumsum())
 .sort_values(["DataModificacaoObito", "Municipio"])
 .rename(columns={"DataModificacaoObito": "data", 
                  "Municipio": "municipio"})
 .drop(columns="Seq_Row_Number"))

mortes
```

```python
serie_historica = confirmados.merge(mortes, on=["municipio", "data"])
serie_historica
```
